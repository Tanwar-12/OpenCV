# Manipulate the pixel values of an image 

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('Image1.jpg')  # Replace with the actual path to your image

# Check if the image is successfully loaded
if image is not None:
    # Accessing and manipulating pixel values
    height, width, channels = image.shape

    # Loop through each pixel
    for y in range(height):
        for x in range(width):
            # Access the RGB values of the pixel
            blue, green, red = image[y, x]

            # Manipulate the pixel values (example: invert colors)
            inverted_pixel = [255 - blue, 255 - green, 255 - red]

            # Update the pixel values
            image[y, x] = inverted_pixel

    # Display the original and manipulated images in Jupyter Notebook
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')

    # Manipulated Image
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Manipulated Image')

    plt.show()

else:
    print("Error: Unable to load the image.")
```

## Intensity Transformation Operations on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('Image1.jpg', cv2.IMREAD_GRAYSCALE)  # Replace with the actual path to your image

# Check if the image is successfully loaded
if image is not None:
    # Apply contrast stretching
    min_intensity = np.min(image)
    max_intensity = np.max(image)
    stretched_image = 255 * (image - min_intensity) / (max_intensity - min_intensity)

    # Display the original and stretched images
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    # Stretched Image
    plt.subplot(1, 2, 2)
    plt.imshow(stretched_image, cmap='gray')
    plt.title('Contrast Stretched Image')

    plt.show()

else:
    print("Error: Unable to load the image.")
```

# Image Enhancement Techniques

## ImageOps.equalize() method

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Open the image
image_path = 'Image1.jpg'  # Replace with the actual path to your image
original_image = Image.open(image_path)

# Check if the image is successfully loaded
if original_image is not None:
    # Perform histogram equalization
    equalized_image = ImageOps.equalize(original_image)

    # Display the original and equalized images
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.title('Original Image')

    # Equalized Image
    plt.subplot(1, 2, 2)
    plt.imshow(equalized_image)
    plt.title('Equalized Image')

    plt.show()

else:
    print("Error: Unable to load the image.")
```

## ImageEnhance.Brightness() and ImageEnhance.Sharpness() method

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

# Open the image
image_path = 'Image1.jpg'  # Replace with the actual path to your image
original_image = Image.open(image_path)

# Check if the image is successfully loaded
if original_image is not None:
    # Enhance brightness
    brightness_factor = 1.5  # Adjust as needed, e.g., 1.0 for no change
    enhanced_brightness_image = ImageEnhance.Brightness(original_image).enhance(brightness_factor)

    # Enhance sharpness
    sharpness_factor = 2.0  # Adjust as needed, e.g., 1.0 for no change
    enhanced_sharpness_image = ImageEnhance.Sharpness(original_image).enhance(sharpness_factor)

    # Display the original, enhanced brightness, and enhanced sharpness images
    plt.figure(figsize=(18, 6))

    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image)
    plt.title('Original Image')

    # Enhanced Brightness Image
    plt.subplot(1, 3, 2)
    plt.imshow(enhanced_brightness_image)
    plt.title('Enhanced Brightness')

    # Enhanced Sharpness Image
    plt.subplot(1, 3, 3)
    plt.imshow(enhanced_sharpness_image)
    plt.title('Enhanced Sharpness')

    plt.show()

else:
    print("Error: Unable to load the image.")
```

# Working with Animated image

## Create and save animated GIF

```python
from PIL import Image, ImageDraw
import os

# Directory to save the GIF
output_directory = 'output_gif'
os.makedirs(output_directory, exist_ok=True)

# Create a sequence of images
num_frames = 10
image_size = (200, 200)

# Generate frames with a moving square
frames = []
for i in range(num_frames):
    img = Image.new('RGB', image_size, color='white')
    draw = ImageDraw.Draw(img)
    square_size = 20 + i * 10
    draw.rectangle([50, 50, 50 + square_size, 50 + square_size], fill='blue')
    frames.append(img)

# Save the frames as an animated GIF
output_path = os.path.join(output_directory, 'animated.gif')
frames[0].save(output_path, save_all=True, append_images=frames[1:], duration=200, loop=0)
print(f"Animated GIF saved to {output_path}")
```
