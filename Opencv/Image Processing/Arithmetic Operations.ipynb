#  Arithmetic Operations on Images using OpenCV  (Addition and Subtraction)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plotting in Jupyter Notebook
%matplotlib inline

# Load two images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Resize images to the same dimensions (optional)
image1 = cv2.resize(image1, (300, 300))
image2 = cv2.resize(image2, (300, 300))

# Add images
result_addition = cv2.add(image1, image2)

# Subtract images
result_subtraction = cv2.subtract(image1, image2)

# Convert BGR to RGB for displaying with matplotlib
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
result_addition_rgb = cv2.cvtColor(result_addition, cv2.COLOR_BGR2RGB)
result_subtraction_rgb = cv2.cvtColor(result_subtraction, cv2.COLOR_BGR2RGB)

# Plot images
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(image1_rgb)
plt.title('SR1')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(image2_rgb)
plt.title('SR2')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(result_addition_rgb)
plt.title('Addition')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(result_subtraction_rgb)
plt.title('Subtraction')
plt.axis('off')

plt.show()
```
