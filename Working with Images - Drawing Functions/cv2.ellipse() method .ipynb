# **Python OpenCV | cv2.ellipse() method**

```python
# Python program to explain cv2.ellipse() method

# importing cv2
import cv2
from matplotlib import pyplot as plt

# path
path = r'D:\Downloads\ES.jpg'

# Reading an image in default mode
image = cv2.imread(path)

# Window name in which image is displayed
window_name = 'Image'

center_coordinates = (120, 100)

axesLength = (100, 50)

angle = 0

startAngle = 0

endAngle = 360

# Red color in BGR
color = (0, 0, 255)

# Line thickness of 5 px
thickness = 5

# Using cv2.ellipse() method
# Draw an ellipse with red line borders of thickness of 5 px
image = cv2.ellipse(image, center_coordinates, axesLength,
                    angle, startAngle, endAngle, color, thickness)

# Convert image from BGR to RGB (matplotlib expects RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Displaying the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis
plt.show()
```
