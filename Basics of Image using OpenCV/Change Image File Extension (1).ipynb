```python
from PIL import Image
import os

# Open the image from a different directory
image_path = r"D:\ST\Pictures\Spider.jpg"
img = Image.open(image_path)

# Save as PNG
output_path_png = os.path.join(os.path.expanduser("~"), "Image.png")
img.save(output_path_png, format="PNG")

# Save as JPG
output_path_jpg = os.path.join(os.path.expanduser("~"), "Image.jpg")
img.save(output_path_jpg, format="JPEG")

# Save as GIF
output_path_gif = os.path.join(os.path.expanduser("~"), "Image.gif")
img.save(output_path_gif, format="GIF")

# Display the new images in Jupyter Notebook
display.display(img)
```

```python
from PIL import Image
import IPython.display as display
import os

# Specify the path to the image
image_path = r"D:\ST\Pictures\Spider.png"

# Open the image
img = Image.open(image_path)

# Save the image with different extensions
extensions = ["PNG", "JPEG", "BMP", "GIF"]

for ext in extensions:
    # Create output path for each extension
    output_path = os.path.join(os.path.expanduser("~"), f"Image_converted.{ext.lower()}")
    
    # Save the image with the current extension
    img.save(output_path, format=ext)
    
    # Display the new image in Jupyter Notebook
    display.display(img)

    # Print the path of the saved image
    print(f"Image converted and saved to: {output_path}")
```
