# Read & Show Image

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Convert BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```

# Change the image color

```python
# Read the image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Set the green channel to 0
image_hide_green = image.copy()
image_hide_green[:, :, 1] = 0  # 1 corresponds to the green channel

# Display the original and modified images side by side
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Image with green channel hidden
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_hide_green, cv2.COLOR_BGR2RGB))
plt.title('Image with Hidden Green Channel')
plt.axis('off')

plt.show()
```

```python
# Read the image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Convert BGR to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original and grayscale images side by side
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray', vmin=0, vmax=255)
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time

# Function to create an image with only the specified channel
def create_channel_image(original_image, channel_idx):
    channel_image = np.zeros_like(original_image)
    channel_image[:, :, channel_idx] = original_image[:, :, channel_idx]
    return channel_image

# Read the original image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Create images with only the red, green, and blue channels
red_channel_image = create_channel_image(image, 2)
green_channel_image = create_channel_image(image, 1)
blue_channel_image = create_channel_image(image, 0)

# List of images to display in the slideshow
image_list = [image, red_channel_image, green_channel_image, blue_channel_image]

# Display the slideshow
plt.figure(figsize=(10, 5))

for idx, img in enumerate(image_list, start=1):
    plt.subplot(1, len(image_list), idx)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {idx}')
    plt.axis('off')
    plt.show(block=False)
    plt.pause(2)  # Display each image for 2 seconds
    plt.clf()  # Clear the figure for the next image

plt.show()
```

# 4 images in one window

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Function to create an image with only the specified channel
def create_channel_image(original_image, channel_idx):
    channel_image = np.zeros_like(original_image)
    channel_image[:, :, channel_idx] = original_image[:, :, channel_idx]
    return channel_image

# Read the original image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Create images with only the red, green, and blue channels
red_channel_image = create_channel_image(image, 2)
green_channel_image = create_channel_image(image, 1)
blue_channel_image = create_channel_image(image, 0)

# Display the images in a 2x2 grid
plt.figure(figsize=(10, 8))

# Original image
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Image with only the red channel
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(red_channel_image, cv2.COLOR_BGR2RGB))
plt.title('Red Channel')
plt.axis('off')

# Image with only the green channel
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(green_channel_image, cv2.COLOR_BGR2RGB))
plt.title('Green Channel')
plt.axis('off')

# Image with only the blue channel
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(blue_channel_image, cv2.COLOR_BGR2RGB))
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```

# Display the same image in all four quadrants

```python
import cv2
import matplotlib.pyplot as plt

# Read the original image
image_path = 'Nature.jpg'  # Replace with the path to your image file
image = cv2.imread(image_path)

# Display the same image in all four quadrants
plt.figure(figsize=(10, 8))

# Original image in the top-left quadrant
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Quadrant 1')
plt.axis('off')

# Original image in the top-right quadrant
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Quadrant 2')
plt.axis('off')

# Original image in the bottom-left quadrant
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Quadrant 3')
plt.axis('off')

# Original image in the bottom-right quadrant
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Quadrant 4')
plt.axis('off')

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```
