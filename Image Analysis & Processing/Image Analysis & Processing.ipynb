## Edge Detection using Pillow

```python
from PIL import Image, ImageFilter
from IPython.display import display

# Opening the image (R prefixed to string
# in order to deal with '\' in paths)
image_path = "Sample.jpg"
image = Image.open(image_path)

# Converting the image to grayscale, as edge detection 
# requires input image to be of mode = Grayscale (L)
image = image.convert("L")

# Detecting Edges on the Image using the argument ImageFilter.FIND_EDGES
image_edges = image.filter(ImageFilter.FIND_EDGES)

# Saving the Image Under the name Edge_Sample.png
edge_image_path = "Edge_Sample.jpg"
image_edges.save(edge_image_path)

# Display the original and edge-detected images in the notebook
display(image)
display(image_edges)
```

```python
from PIL import Image, ImageFilter
from IPython.display import display

img_path = "Sample.jpg"

# Open the image
img = Image.open(img_path)

# Convert the image to grayscale
img_gray = img.convert("L")

# Calculate edges using the Sobel Operator
final_img = img_gray.filter(ImageFilter.Kernel((3, 3), (-1, -1, -1, -1, 8, -1, -1, -1, -1), 1, 0))

# Save the edge-detected image
edge_img_path = "EDGE_Sample.jpg"
final_img.save(edge_img_path)

# Display the original and edge-detected images in the notebook
display(img)
display(final_img)
```

## Overlay an image on another image in Python

```python
from PIL import Image
from IPython.display import display

# Opening the primary image (used in background)
img1 = Image.open("Bg.jpg")

# Opening the secondary image (overlay image)
img2 = Image.open("k.png")

# Pasting img2 image on top of img1 starting at coordinates (0, 0)
img1.paste(img2, (0, 0), mask=img2)

# Displaying the image in the notebook
display(img1)
```

## Create transparent png image with Python  Pillow

```python
from PIL import Image
from IPython.display import display

# Open the original image
img_path = "SAKSHI TANWAR .png"
img = Image.open(img_path)

# Convert black pixels to transparent
rgba = img.convert("RGBA")
datas = rgba.getdata()
newData = [(255, 255, 255, 0) if item[0] == 0 and item[1] == 0 and item[2] == 0 else item for item in datas]
rgba.putdata(newData)

# Save the transparent image
transparent_img_path = "transparent_image.png"
rgba.save(transparent_img_path, "PNG")

# Display the original and transparent images in the notebook
display(img)
display(rgba)
```
